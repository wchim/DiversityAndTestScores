```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('17-18DemoMath.csv')
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
sns.barplot(x=df['Total Enrollment'].sort_values(ascending=False).head(10),
            y=df.sort_values('Total Enrollment', ascending=False)['School Name'].head(10),
            color='gray',
            alpha=1,)
    
plt.title("Top 10 Schools by Total Student Enrollment during 2017-18 School Year")
plt.xlabel("Total Student Enrollment")
plt.ylabel("School Name")
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10,5))
sns.set_style("whitegrid")

weights_m = np.ones_like(np.array(df['% Male']))/float(len(np.array(df['% Male'])))
axs[0].hist(df['% Male'], 
            weights=weights_m,
            bins=100,
            color='lightblue')

weights_f = np.ones_like(np.array(df['% Female']))/float(len(np.array(df['% Female'])))
axs[1].hist(df['% Female'],
            weights=weights_f,
            bins=100,
            color='pink')

fig.suptitle("Gender Distribution Across 1105 Schools During the 2017-18 School Year")
axs[0].legend(["Male"])
axs[0].set_xlabel("Percentage of Male Students")
axs[0].set_ylabel("Percentage of Schools")
axs[1].legend(["Female"])
axs[1].set_xlabel("Percentage of Female Students")
fig.show()
```

```python
fig, axs = plt.subplots(2, 2, sharex = True, sharey = True, figsize=(10,10))
sns.set_style("whitegrid")

weights_a = np.ones_like(np.array(df['% Asian']))/float(len(np.array(df['% Asian'])))
axs[0][0].hist(df['% Asian'],
               weights=weights_a,
               bins=100,
               color='orange')

weights_b = np.ones_like(np.array(df['% Black']))/float(len(np.array(df['% Black'])))
axs[0][1].hist(df['% Black'],
               weights=weights_b,
               bins=100,
               color='purple')

weights_h = np.ones_like(np.array(df['% Hispanic']))/float(len(np.array(df['% Hispanic'])))
axs[1][0].hist(df['% Hispanic'],
               weights=weights_h,
               bins=100,
               color='green')

weights_w = np.ones_like(np.array(df['% White']))/float(len(np.array(df['% White'])))
axs[1][1].hist(df['% White'],
               weights=weights_w,
               bins=100,
               color='gray')

fig.suptitle("Ethnicity Distribution Across 1105 Schools During the 2017-18 School Year")
fig.text(0.5, 0.04, "Percentage of Students by Ethnicity", ha="center")
fig.text(0.04, 0.5, "Percentage of Schools", va="center", rotation="vertical")

axs[0][0].legend(["Asian"])
axs[0][0].set_xlabel("")
axs[0][1].legend(["Black"])
axs[0][1].set_xlabel("")
axs[1][0].legend(["Hispanic"])
axs[1][0].set_xlabel("")
axs[1][1].legend(["White"])
fig.show()
```
