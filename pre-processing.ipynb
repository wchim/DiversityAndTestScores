```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
math_df = pd.read_csv('2013_-2018_School_Math_Results.csv')
math_temp = math_df.copy()
math_temp.head()
```

```python
math_temp.columns
```

```python
math_temp.drop(math_temp[math_temp.Year != 2018].index,
               inplace=True)
math_temp.drop(math_temp[math_temp.Grade != 'All Grades'].index,
               inplace=True)
math_temp.drop(math_temp[math_temp.Category != 'All Students'].index,
               inplace=True)
math_temp.drop(['Grade', 'Year', 'Category'], axis=1, inplace=True)
math_temp.head()
```

```python
print('Math Scores Dataset Dimensions')
print('Original Dimension:{}\nProcessed Dimension:{}'.format(math_df.shape, math_temp.shape))
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10,5))
sns.heatmap(math_df.isnull(), cbar=False, ax=axs[0])
sns.heatmap(math_temp.isnull(), cbar=False, ax=axs[1])
```

```python
demo_df = pd.read_csv('2018-2019_School_Demographic_Snapshot.csv')
demo_temp = demo_df.copy()
demo_temp.head()
```

```python
demo_temp.columns
```

```python
demo_temp.drop(demo_temp[demo_temp.Year != '2017-18'].index, inplace=True)
demo_temp.drop(['School Name', 'Year'], axis=1, inplace=True)
demo_temp.head()
```

```python
print('Demographic Dataset Dimensions')
print('Original Dimension:{}\nProcessed Dimension:{}'.format(demo_df.shape, demo_temp.shape))
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10,5))
sns.heatmap(demo_df.isnull(), cbar=False, ax=axs[0])
sns.heatmap(demo_temp.isnull(), cbar=False, ax=axs[1])
```

```python
main_df = pd.merge(left=math_temp, right=demo_temp, how='inner', on='DBN')
main_df.rename(columns={'# Multiple Race Categories Not Represented': '# Others',
              '% Multiple Race Categories Not Represented': '% Others'},
             inplace=True)
main_df.head()
```

```python
print('Merged Dataset Dimension')
print('Original Dimension:{}'.format(main_df.shape))
```
